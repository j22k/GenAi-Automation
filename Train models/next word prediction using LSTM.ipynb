```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('/teamspace/studios/this_studio/medical_tc_train.csv')
df.head(3)
```

```python
df = df['medical_abstract']
```

```python
df
```

```python
df.shape
```

```python

df = df.head(1000)
df
```

```python
text = df.to_list()
```

```python
text
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(text)
seq = tokenizer.texts_to_sequences(text)
```

```python
seq[:10]
```

```python
tokenizer.word_index
```

```python
X = []
y = []
total_words_dropped = 0

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
            X.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped += 1

print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X[:10]
```

```python
y[:10]
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(X)
```

```python
X
```

```python
X.shape
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
vocab_size
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(200, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.004),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=45)
```

```python
model.save('LSTM.h5')
```

```python
import pickle
# Save tokenizer
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
vocab_array = np.array(list(tokenizer.word_index.keys()))
```

```python
vocab_array
```

```python
def make_prediction(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
        prediction = str(vocab_array[prediction - 1])
        print(vocab_array[np.argsort(model.predict(text_padded)) - 1].ravel()[:-3])
        text += " " + prediction
    return text
```

```python
make_prediction("cloudy", 5)
```

```python
import matplotlib.pyplot as plt

# Plotting the loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(model.history['loss'], label='Training Loss')
plt.plot(model.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plotting the accuracy
plt.subplot(1, 2, 2)
plt.plot(model.history['accuracy'], label='Training Accuracy')
plt.plot(model.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
def plot_graphs(history,string):
    plt.plot(history.history[string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.show()
plot_graphs(model, 'accuracy')
plot_graphs(model, 'accuracy')
```
